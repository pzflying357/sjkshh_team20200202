```python
# -*- coding: utf-8 -*-
import numpy as np
import pandas as pd
from pandas import Series
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.charts import Pie
from pyecharts.charts import Funnel, Page
from pyecharts.globals import ThemeType

#ex1
f=open('D:/pystudy/数据可视化小组/task_4_lianjia_sale.csv',encoding='utf-8')
lianjia_saledata = pd.read_csv(f)

f.close

print(lianjia_saledata.head())
fy_avg = [('区域平均价格','mean')]
df1 = lianjia_saledata.groupby('区域')['价格'].agg(fy_avg)

df_avg=df1.sort_values(by="区域平均价格")
print(df_avg.head())

#ex2

x=df_avg.index.values.tolist()
y=df_avg['区域平均价格'].values.tolist()
y1 = [int(i) for i in y]

bar = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))
    .add_xaxis(x)
    .add_yaxis('均价',y1)
    .set_global_opts(title_opts=opts.TitleOpts(title="上海各区域平均价格"))
)
bar.render_notebook() #直接在jupytr notebook中渲染

#3
fy_count = [('房源数量','count')]
df_zhuangxiu = lianjia_saledata.groupby(['装修类型','区域'])['区域'].agg(fy_count)
#print(df_zhuangxiu.head(30))

#4

v1=df_zhuangxiu.index.values.tolist()


v2=df_zhuangxiu['房源数量'].values.tolist()
data=[list(z) for z in zip(v1,v2)]
for i in range(15):
    data2=data[i:-1:15]
    #print(data2)
c = (
    Pie()
    .add("",data2 )
    .set_global_opts(title_opts=opts.TitleOpts(title="装修类型占比情况"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c} --{d}%"))#格式化标签输出内容
)

c.render_notebook()

```
